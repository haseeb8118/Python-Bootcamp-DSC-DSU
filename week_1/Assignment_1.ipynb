# Q1: Define a function to print a string diagonally

```python
def diagonal_print():
  words = input("Enter your name: ")
  space = ""
  for char in words:
    print(space, char)
    space += " "

diagonal_print()
```

# Q2: Create a program to take as input 5 student records in the following format:
**roll_num** | **name** | **age** | **marks**(out of 100)

And then output the records in a tabular form with class average, class highest and class lowest at end in the following format.

Use dictionaries (list of dictionaries in exact)

Insert atleast 5 records

Input must be user-given

(Optional) validate the user input, i.e marks aren't greater 100 and other such validations you think there might be

```python
print("\n------------------------------------------------------------------------------------------\n")

first = [int(input("Roll Number: ")), input("Name: "), int(input("Age: ")), int(input("Marks: "))]

while first[3] > 100:
        print("Please enter no. between 1 to 100")
        first[3] = int(input("Marks: "))
        if first[3] < 100:           
         break

print("\n------------------------------------------------------------------------------------------\n")

second = [int(input("Roll Number: ")), input("Name: "), int(input("Age: ")), int(input("Marks: "))]

while second[3] > 100:
        print("Please enter no. between 1 to 100")
        second[3] = int(input("Marks: "))
        if second[3] < 100:           
         break

print("\n------------------------------------------------------------------------------------------\n")

third = [int(input("Roll Number: ")), input("Name: "), int(input("Age: ")), int(input("Marks: "))]

while third[3] > 100:
        print("Please enter no. between 1 to 100")
        third[3] = int(input("Marks: "))
        if third[3] < 100:           
         break

print("\n------------------------------------------------------------------------------------------\n")

fourth = [int(input("Roll Number: ")), input("Name: "), int(input("Age: ")), int(input("Marks: "))]

while fourth[3] > 100:
        print("Please enter no. between 1 to 100")
        fourth[3] = int(input("Marks: "))
        if fourth[3] < 100:           
         break

print("\n------------------------------------------------------------------------------------------\n")

fifth = [int(input("Roll Number: ")), input("Name: "), int(input("Age: ")), int(input("Marks: "))]
        
while fifth[3] > 100:
        print("Please enter no. between 1 to 100")
        fifth[3] = int(input("Marks: "))
        if fifth[3] < 100:           
         break

print("\n------------------------------------------------------------------------------------------\n")

from tabulate import tabulate

print(tabulate([[first[0], first[1],first[2], first[3]],
                [second[0], second[1],second[2], second[3]],
                [third[0], third[1], third[2], third[3]],
                [fourth[0], fourth[1],fourth[2], fourth[3]],
                [fifth[0], fifth[1], fifth[2], fifth[3]]],
                headers=["Roll_no", "Name", "Age", "Marks"]))
```
